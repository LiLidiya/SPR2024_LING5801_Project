```python
import nltk
import re
```

# Data Modeling Frequency Distribution

# Corpus for each rating


```python
onerate = open("oneRate.txt", "r").readlines()
tworate = open("twoRate.txt", "r").readlines()
threerate = open("threeRate.txt","r").readlines()
fourrate= open("fourRate.txt","r").readlines()
fiverate = open("fiveRate.txt","r").readlines()

oneCorpus = []
for line in onerate:
    line = re.sub(r'^[\d\w_]+\s+', '', line).strip()  # remove audio file info
    line = re.sub(r'<[\w\']+>', '', line)             # remove verbal repairs
    line = re.sub(r'\w+-', '', line)                  # remove fragments
    line = re.sub(r'\*([\w\']+)\*', '\1', line)       # remove mispronunciation markers
    oneCorpus.append(line.split())  

twoCorpus = []
for line in tworate:
    line = re.sub(r'^[\d\w_]+\s+', '', line).strip()  # remove audio file info
    line = re.sub(r'<[\w\']+>', '', line)             # remove verbal repairs
    line = re.sub(r'\w+-', '', line)                  # remove fragments
    line = re.sub(r'\*([\w\']+)\*', '\1', line)       # remove mispronunciation markers
    twoCorpus.append(line.split())    
    
threeCorpus = []
for line in threerate:
    line = re.sub(r'^[\d\w_]+\s+', '', line).strip()  # remove audio file info
    line = re.sub(r'<[\w\']+>', '', line)             # remove verbal repairs
    line = re.sub(r'\w+-', '', line)                  # remove fragments
    line = re.sub(r'\*([\w\']+)\*', '\1', line)       # remove mispronunciation markers
    threeCorpus.append(line.split())  
    
fourCorpus = []
for line in fourrate:
    line = re.sub(r'^[\d\w_]+\s+', '', line).strip()  # remove audio file info
    line = re.sub(r'<[\w\']+>', '', line)             # remove verbal repairs
    line = re.sub(r'\w+-', '', line)                  # remove fragments
    line = re.sub(r'\*([\w\']+)\*', '\1', line)       # remove mispronunciation markers
    fourCorpus.append(line.split())    
    
fiveCorpus = []
for line in fiverate:
    line = re.sub(r'^[\d\w_]+\s+', '', line).strip()  # remove audio file info
    line = re.sub(r'<[\w\']+>', '', line)             # remove verbal repairs
    line = re.sub(r'\w+-', '', line)                  # remove fragments
    line = re.sub(r'\*([\w\']+)\*', '\1', line)       # remove mispronunciation markers
    fiveCorpus.append(line.split())    
    

    
```

### Task 1: Unigram and bigram MLE language model

Using NLTK's `FreqDist()` and `ConditionalFreqDist()` classes train unigram and bigram language models based on the rating corpus.  

```python
print("-------------One Rate----------------")
unigramOne = nltk.FreqDist()

for sentence in oneCorpus:
    for word in sentence:
        unigramOne[word] += 1

print(unigramOne.most_common(100))

print("-------------Two Rate----------------")
unigramTwo = nltk.FreqDist()

for sentence in twoCorpus:
    for word in sentence:
        unigramTwo[word] += 1

print(unigramTwo.most_common(100))
print("-------------Three Rate----------------")

unigramThree = nltk.FreqDist()

for sentence in threeCorpus:
    for word in sentence:
        unigramThree[word] += 1

print(unigramThree.most_common(100))
print("-------------Four Rate----------------")

unigramFour = nltk.FreqDist()

for sentence in fourCorpus:
    for word in sentence:
        unigramFour[word] += 1

print(unigramFour.most_common(100))

print("-------------Five Rate----------------")

unigramFive = nltk.FreqDist()

for sentence in fiveCorpus:
    for word in sentence:
        unigramFive[word] += 1

print(unigramFive.most_common(100))
print("-----------------------------")
```

```python
bigramOne = nltk.ConditionalFreqDist()
bigramTwo = nltk.ConditionalFreqDist()
bigramThree = nltk.ConditionalFreqDist()
bigramFour = nltk.ConditionalFreqDist()
bigramFive = nltk.ConditionalFreqDist()

for sentence in oneCorpus:
    preceding = "<s>"
    for word in sentence:
        bigramOne[preceding][word] += 1
        preceding = word
    bigramOne[preceding]["</s>"] += 1

for sen in twoCorpus:
    preceding = "<s>"
    for word in sen:
        bigramTwo[preceding][word] += 1
        preceding = word
    bigramTwo[preceding]["</s>"] += 1

for sentence in threeCorpus:
    preceding = "<s>"
    for word in sentence:
        bigramThree[preceding][word] += 1
        preceding = word
    bigramThree[preceding]["</s>"] += 1

for sentence in fourCorpus:
    preceding = "<s>"
    for word in sentence:
        bigramFour[preceding][word] += 1
        preceding = word
    bigramFour[preceding]["</s>"] += 1
    
for sentence in fiveCorpus:
    preceding = "<s>"
    for word in sentence:
        bigramFive[preceding][word] += 1
        preceding = word
    bigramFive[preceding]["</s>"] += 1

# bigram["not"].most_common() ##################################

most = unigramOne.most_common(100)
# for i in most:
#     print(i)
#     print(bigram[i[0]].most_common())
#     print("-------------------------------------------------")
print(bigramOne['extremely'].most_common())
print("-------------------------------------------------")
print(bigramTwo['extremely'].most_common())
print("-------------------------------------------------")
print(bigramThree['extremely'].most_common())
print("-------------------------------------------------")
print(bigramFour['extremely'].most_common())
print("-------------------------------------------------")
print(bigramFive['extremely'].most_common())
print("-------------------------------------------------")

```
